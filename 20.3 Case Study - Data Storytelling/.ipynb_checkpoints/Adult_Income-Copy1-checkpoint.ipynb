### Data Story Telling Case Study

## Goal 


This case study is to explore and analyze the adult income data and provide insights.



## Questions of Interest

1. What is range of the income in different countries by gender?
2. Does the income vary based on Race?
3. Do Males get paid more than Females?
4. Does income vary based on age group?
5. What is relation between education and income?


## Data Exploration 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline



#Create label for price
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
#used for expanding the no o viible columns of dataframe
```

```python
df = pd.read_csv('adult.csv')
```

```python
df.info()
```

```python
sns.catplot(y='age',x='income',data=df, hue="gender", kind="point", )

plt.title("Age Vs Income")
plt.xticks(rotation=90 )

plt.show()
```

```python
df.drop(df[df['native-country'] == '?'].index, inplace=True)
df.drop(df[df['occupation'] == '?'].index, inplace=True)
```

```python
df_50up = df[df.income == '>50K']
df_50 = df[df.income == '<=50K']
```

```python
#df_50up_cnt.sort_values(by='counts',ascending=False)
```

```python
df_50_cnt=df_50[['native-country','income']].groupby('native-country').size().reset_index(name='counts')
df_50_cnt = df_50_cnt.sort_values(by='counts',ascending=False)
sns.scatterplot(x='native-country',y='counts',data=df_50_cnt )
plt.title("Count of people earning income less than 50K")
plt.xticks(rotation=90)
plt.show()
```

```python
df_50up_cnt=df_50up[['native-country','income']].groupby('native-country').size().reset_index(name='counts')
df_50up_cnt=df_50up_cnt.sort_values(by='counts',ascending=False)
sns.scatterplot(x='native-country',y='counts',data=df_50up_cnt )
plt.title("Count of people earning income more than 50K")
plt.xticks(rotation=90)
plt.show()
```

```python


df_50occ=df_50[['occupation','income']].groupby('occupation').size().reset_index(name='counts')
df_50occ = df_50occ.sort_values(by='counts',ascending=False)
sns.catplot(x='occupation',y='counts',data=df_50occ, kind='bar' )
plt.title("Occupation Vs Income < 50K")
plt.xticks(rotation=90)
plt.show()
```

```python
df_50occ=df_50[['occupation','income']].groupby('occupation').size().reset_index(name='counts')
df_50occ = df_50occ.sort_values(by='counts',ascending=False)

df_50upocc=df_50up[['occupation','income']].groupby('occupation').size().reset_index(name='counts')
df_50upocc = df_50upocc.sort_values(by='counts',ascending=False)

labels=[1,2]
fig, axes= plt.subplots(1,2, figsize=(15,5))

axes[0].bar(df_50occ['occupation'],df_50occ['counts'] )

axes[0].set_title("Occupation Vs Income < 50K")
#axes[0].set_xticklabels(labels, rotation=90)
axes[0].tick_params(labelrotation=90)
axes[0].set_xlabel('Occupation')
axes[0].set_ylabel('Counts')

axes[1].bar(df_50upocc['occupation'],df_50upocc['counts'] )
axes[1].set_title("Occupation Vs Income > 50K")
axes[1].tick_params(labelrotation=90)
axes[1].set_xlabel('Occupation')
axes[1].set_ylabel('Counts')


plt.show()



```
