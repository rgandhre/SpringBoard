## Problem Statement:

## Questions of Interest

    . Does income vary based on age group?
    . Do Males get paid more than Females?
    . What is relation between education and income?
    . Does the income vary based on Race?

## Data Exploration 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline



#Create label for price
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
#used for expanding the no o viible columns of dataframe
```

```python
df = pd.read_csv('adult.csv')
```

```python
df.info()
```

```python
df.drop(df[df['native-country'] == '?'].index, inplace=True)
df.drop(df[df['occupation'] == '?'].index, inplace=True)
```

```python
df_50up = df[df.income == '>50K']
df_50 = df[df.income == '<=50K']
```

## Do people earn more as they grow older?

Yes!!  You gain more experience as you grow older and tend to earn more.

Looks like Male earn more than Females

```python
sns.catplot(x="income", y="age", hue='gender', kind="point",
            palette="pastel", edgecolor=".6",
            data=df
           )
plt.xlabel("Income Category")
plt.title("Income VS Age")
plt.show()
```

## Is there a gender income gap?

Females are paid less than Males

```python
sns.catplot(x="income", hue="gender", kind="count",
            palette="pastel", edgecolor=".6",
            data=df
           )
plt.xlabel("Income Category")
plt.title("Income VS Gender")
plt.show()
```

## Is the income gap because of education?

. People with Bachelors degreee are paid >50K 

. Less Female members hold Bachelors degree

```python
sns.catplot(x="income", hue="education",
            col="gender", palette='colorblind', sharey=False,
            kind="count", data=df)

plt.show()
```

## Is there an income race gap?

. Based on the data provided, Race does play a role, it shows'White Males' tend to earn more

```python
sns.catplot(x="income", hue="race",
            col="gender", palette='colorblind',
            kind="count", data=df)
plt.show()
```

```python

sns.catplot(x="income", hue="education",
            col="race", palette='colorblind',
            kind="count", data=df)

plt.show()
```

 ## Conclusion
