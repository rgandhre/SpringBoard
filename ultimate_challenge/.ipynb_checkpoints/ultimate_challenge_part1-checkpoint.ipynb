# Part 1 ‐ Exploratory data analysis
The attached logins.json file contains (simulated) timestamps of user logins in a particular
geographic location. Aggregate these login counts based on 15minute time intervals, and
visualize and describe the resulting time series of login counts in ways that best characterize the
underlying patterns of the demand. Please report/illustrate important features of the demand,
such as daily cycles. If there are data quality issues, please report them.

```python
df_login = pd.read_json('logins.json')
df_login
```

```python
df_login.info()
```

```python
df_login.resample('15T', on='login_time').count()
#Resample on 15 minute time interval and plot
plt.figure(figsize=(15,10))
plt.plot(df_login.resample('15T', on='login_time').count())
plt.show()
```

```python
#df_login[(df_login['login_time'] > '1970-01-01 15:00:00') & (df_login['login_time']  < '1970-01-01 20:30:00')]
#df_login[(df_login['login_time'] > '1970-01-01 20:00:00') & (df_login['login_time']  < '1970-01-01 20:15:00')]
                                                                                                                          
```

After plotting the 15 minute interval time, it shows that there is spike in the month of March

```python
#Resample on monthly time interval
df_monthly = df_login.resample('M', on='login_time').count()
df_monthly
```

```python
plt.figure(figsize=(10,8))
plt.plot(df_monthly)
plt.xticks(rotation=90)
plt.show()
```

```python
format = '%Y-%m-%d'
#Split data for each months

df_Jan = df_login[(df_login['login_time'].dt.strftime(format) >= '1970-01-01') 
                    & (df_login['login_time'].dt.strftime(format) <= '1970-01-31')]

df_Feb = df_login[(df_login['login_time'].dt.strftime(format) >= '1970-02-01') 
                    & (df_login['login_time'].dt.strftime(format) <= '1970-02-28')]

df_Mar = df_login[(df_login['login_time'].dt.strftime(format) >= '1970-03-01') 
                    & (df_login['login_time'].dt.strftime(format) <= '1970-03-31')]


df_Apr = df_login[(df_login['login_time'].dt.strftime(format) >= '1970-04-01') 
                    & (df_login['login_time'].dt.strftime(format) <= '1970-04-30')]
```

```python
#Plot Daily Data for all months
figure, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2,figsize=(20,10))

plt.suptitle("Daily Data for all Months")
ax1.plot(df_Jan.resample('D', on='login_time').count())
ax1.set_title('January Daily Data')
ax1.tick_params(axis="both", direction="out", pad=15, rotation=90)

ax2.plot(df_Feb.resample('D', on='login_time').count())
ax2.set_title('February Daily Data')
ax2.tick_params(axis="both", direction="out", pad=15, rotation=90)


ax3.plot(df_Mar.resample('D', on='login_time').count())
ax3.set_title('March Daily Data')
ax3.tick_params(axis="both", direction="out", pad=15, rotation=90)

ax4.plot(df_Apr.resample('D', on='login_time').count())
ax4.set_title('April Daily Data')
ax4.tick_params(axis="both", direction="out", pad=15, rotation=90)

figure.tight_layout()
plt.show()

```

Looking at the graphs for each month, March had spike in login counts and April showed decline in the login counts.
